```python
import src.config as config
from src.data import dowload_one_file_of_raw_data
from datetime import datetime, timedelta

# get end-date as current date because we want the most recent data when we run pipeline
# get start-date as 2 years back because api limits the amount of data we can get.
end_date = datetime.now().date().strftime('%Y-%m-%d')

end_date_datetime = datetime.strptime(end_date, '%Y-%m-%d')

start_date_datetime = end_date_datetime - timedelta(days=2*365)

start_date = start_date_datetime.strftime('%Y-%m-%d')

prices = dowload_one_file_of_raw_data(symbol='AAPL', start_date=start_date, end_date=end_date) 
print(start_date, end_date)
prices
```
